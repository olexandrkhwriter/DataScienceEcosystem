```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO  # To handle the FutureWarning

# Function to scrape GameStop revenue data
def scrape_gamestop_revenue(url):
    try:
        # Send a request to the URL
        headers = {"User-Agent": "Mozilla/5.0"}
        response = requests.get(url, headers=headers)

        # Check if the response is successful
        if response.status_code != 200:
            print(f"Failed to retrieve page. Status code: {response.status_code}")
            return None

        # Parse the page content
        soup = BeautifulSoup(response.content, "html.parser")

        # Find all the tables on the page
        tables = soup.find_all("table")

        # Check if at least two tables are found
        if len(tables) < 2:
            print("Revenue table not found on the page.")
            return None

        # Extract the second table (which contains revenue data)
        gamestop_revenue_table = tables[1]

        # Convert the table into a pandas DataFrame
        gamestop_revenue_df = pd.read_html(str(gamestop_revenue_table))[0]

        # Print the first few rows to inspect the structure
        print(gamestop_revenue_df.head())

        # Check column names to verify proper access
        print("Column names:", gamestop_revenue_df.columns)

        # Rename columns based on actual names in the DataFrame
        # Example: Update these according to what the output prints
        gamestop_revenue_df = gamestop_revenue_df.rename(columns={
            "GameStop Quarterly Revenue(Millions of US $)": "Date", 
            "GameStop Quarterly Revenue(Millions of US $).1": "Revenue"
        })

        # Clean the revenue column by removing commas and dollar signs, and convert to float
        gamestop_revenue_df["Revenue"] = gamestop_revenue_df["Revenue"].str.replace(",", "").str.replace("$", "").astype(float)
        
        # Drop any rows with missing data
        gamestop_revenue_df = gamestop_revenue_df.dropna()

        return gamestop_revenue_df

    except Exception as e:
        print(f"Error: {e}")
        return None

# URL of GameStop revenue data page
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

# Call the function to scrape data
gamestop_revenue = scrape_gamestop_revenue(url)

# Display the result if the data was successfully retrieved
if gamestop_revenue is not None:
    print("GameStop Revenue Data:")
    print(gamestop_revenue.head())
else:
    print("Failed to retrieve GameStop revenue data.")
```
